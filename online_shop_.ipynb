**Let's know top ten product sales**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# نجيب Top 10 products بالمبيعات
top_products = (
    train_df.groupby("Product Name")
      .agg(total_sales=("Sales","sum"))
      .sort_values("total_sales", ascending=False)
      .head(10)
)

plt.figure(figsize=(12,9))

# المحور الأساسي: المبيعات
ax = sns.barplot(x=top_products.index, y=top_products["total_sales"], color="skyblue")

# كتابة القيم فوق كل عمود (Sales)
for i, sales_val in enumerate(top_products["total_sales"]):
    ax.text(i, sales_val + (sales_val*0.02), f"${sales_val:,.0f}",
            ha="center", va="bottom", fontsize=9, fontweight="bold")

# العناوين
ax.set_title("Top 10 Products by Sales", fontsize=14, fontweight="bold")
ax.set_xlabel("Product Name")
ax.set_ylabel("Total Sales ($)")

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**Top 10 Products by Sales**

This bar chart shows the top 10 best-selling products in terms of total sales revenue. The Canon ImageCLASS 2200 Advanced Copier leads significantly with over $61K in sales, far ahead of the next products. Other high performers include the Cisco Telepresence System EX90 and HON 5400 Series Task Chairs. This visualization highlights which products generate the highest revenue, helping the business identify key items to prioritize in inventory, marketing, and sales strategies.

> Relationship between Delivery Time, State, and Sales



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# تحويل الأعمدة لتواريخ
train_df["Order Date"] = pd.to_datetime(train_df["Order Date"], dayfirst=True)
train_df["Ship Date"] = pd.to_datetime(train_df["Ship Date"], dayfirst=True)

# حساب مدة التوصيل
train_df["delivery_days"] = (train_df["Ship Date"] - train_df["Order Date"]).dt.days

# تجميع البيانات حسب State
region_summary = (
    train_df.groupby("State")
      .agg(avg_delivery=("delivery_days","mean"), total_sales=("Sales","sum"))
      .reset_index()
)

# رسم scatterplot يوضح العلاقة
plt.figure(figsize=(14,6))
sns.scatterplot(data=region_summary, x="avg_delivery", y="total_sales", hue="State", s=200)

# كتابة القيم على النقاط
for i, row in region_summary.iterrows():
    plt.text(row["avg_delivery"]+0.1, row["total_sales"], row["State"], fontsize=9)

plt.title("Relationship between Delivery Time, State, and Sales", fontsize=14, fontweight="bold")
plt.xlabel("Average Delivery Days")
plt.ylabel("Total Sales ($)")
plt.tight_layout()
plt.show()
```

**Relationship between Delivery Time, State, and Sales**

In this scatterplot we note there's States such as new york taxes california and more have high sales and average delivery is more than 3.5 days , One the Other Hand states like louisiana ,virginia ,Rhode island ,north dakota have average dilevery time less  than 3.5 days at the same time lowest sales

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import kagglehub

# تحميل الداتا من KaggleHub
path = kagglehub.dataset_download("rohitsahoo/sales-forecasting")
train_df = pd.read_csv(f'{path}/train.csv')

# تحويل الأعمدة لتواريخ
train_df["Order Date"] = pd.to_datetime(train_df["Order Date"], dayfirst=True)
train_df["Ship Date"] = pd.to_datetime(train_df["Ship Date"], dayfirst=True)

# إضافة عمود للسنة
train_df["Year"] = train_df["Order Date"].dt.year

# إجمالي المبيعات لكل ولاية
state_totals = train_df.groupby("State")["Sales"].sum().sort_values(ascending=False)

# اختيار أعلى 5 ولايات
top10_states = state_totals.head(10).index.tolist()

# حساب المبيعات لكل ولاية من الـ top 5 لكل سنة
state_year_sales = (
    train_df[train_df["State"].isin(top10_states)]
    .groupby(["Year", "State"])["Sales"]
    .sum()
    .reset_index()
)

# الرسم باستخدام seaborn
plt.figure(figsize=(12,6))
sns.barplot(data=state_year_sales, x="Year", y="Sales", hue="State")
plt.title("Top 5 States Sales Over Years")
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.legend(title="State")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import kagglehub

# تحميل الداتا
path = kagglehub.dataset_download("rohitsahoo/sales-forecasting")
train_df = pd.read_csv(f'{path}/train.csv')

# تحويل الأعمدة لتواريخ
train_df["Order Date"] = pd.to_datetime(train_df["Order Date"], dayfirst=True)
train_df["Ship Date"] = pd.to_datetime(train_df["Ship Date"], dayfirst=True)

# إضافة عمود للسنة
train_df["Year"] = train_df["Order Date"].dt.year

# فلترة على ولاية كاليفورنيا فقط
df_california = train_df[train_df["State"] == "California"]

# حساب المبيعات لكل Sub-Category لكل سنة
subcat_sales_california = (
    df_california.groupby(["Year", "Sub-Category"])["Sales"]
    .sum()
    .reset_index()
)

# رسم Bar Plot
plt.figure(figsize=(14,7))
sns.barplot(
    data=subcat_sales_california,
    x="Year",
    y="Sales",
    hue="Sub-Category"
)
plt.title("Sub-Category Sales Over Years in California")
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.legend(title="Sub-Category", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```
